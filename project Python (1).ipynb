```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

counts = data['License_Validity'].value_counts()[['Valid', 'Suspended', 'Expired']]

palette = {
    "Valid": "#5cb85c",       # green
    "Suspended": "#d9534f",   # red
    "Expired": "#f0ad4e"      # orange
}

plt.figure(figsize=(6,5))
bars = plt.bar(counts.index, counts.values,
               color=[palette[x] for x in counts.index],
               edgecolor="black")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + height*0.01,
        f"{int(height)}",
        ha="center",
        fontsize=11
    )


plt.title("Violations by License Status", fontsize=15, fontweight="bold")
plt.xlabel("License Status", fontsize=13)
plt.ylabel("Number of Violations", fontsize=13)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

high_penalty = data[data['Penalty_Points'] >= 4]


state_counts = high_penalty['Registration_State'].value_counts()

plt.figure(figsize=(10,6))
plt.scatter(
    x=state_counts.index,
    y=state_counts.values,
    s=state_counts.values*8,
    color='teal',
    alpha=0.5
)

for i, count in enumerate(state_counts.values):
    plt.text(i, count, str(count), ha='center', va='center', fontsize=10, color='white', fontweight='bold')


plt.xlabel("State")
plt.ylabel("Number of High-Penalty Drivers")
plt.title("High-Penalty Drivers by State")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


road_violation_pivot = data.pivot_table(
    index='Violation_Type',
    columns='Road_Condition',
    aggfunc='size',
    fill_value=0
)

plt.figure(figsize=(10,6))
sns.heatmap(road_violation_pivot, annot=True, fmt='g', cmap='YlGnBu')
plt.title("Frequency of Violations by Road Condition")
plt.xlabel("Road Condition")
plt.ylabel("Violation Type")
plt.show()
```

```python
court_required = data[data['Court_Appearance_Required'] == 'Yes']


violation_counts = court_required['Violation_Type'].value_counts()


plt.figure(figsize=(6,6))
violation_counts.head(5).plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Violations Requiring Court Appearance",fontweight="bold")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
limits = {'Car':4, 'Scooter':2, 'Truck':2, 'Bus':4, 'Auto Rickshaw':3, 'Bike':2}

data['Overloaded'] = data['Number_of_Passengers'] > data['Vehicle_Type'].map(limits)

sns.countplot(x='Vehicle_Type', data=data[data['Overloaded']], color="orange",width=0.20)
plt.title("Number of Overloaded Vehicles by Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Count of Overloaded Vehicles")
plt.show()
```

```python
positive_data = data.loc[data['Breathalyzer_Result'] == "Positive", ['Driver_Age', 'Breathalyzer_Result']]
print(positive_data)
```

```python
import pandas as pd

bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['-18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+']

data['Age'] = pd.cut(data['Driver_Age'], bins=bins, labels=labels, right=False)

positive_df = data[data['Breathalyzer_Result'] == 'Positive']

positive_counts = positive_df['Age'].value_counts().sort_index()
print(positive_counts)


import matplotlib.pyplot as plt

positive_counts.plot(kind='barh')
plt.title("Positive Breathalyzer Results by Age Group",fontweight="bold")
plt.xlabel("Count of Positive Tests")
plt.ylabel("Age Group")
plt.show()


```
